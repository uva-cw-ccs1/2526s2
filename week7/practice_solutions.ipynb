## Part 1: Pandas Recap 🐼

We'll use a music streaming dataset throughout this section.

**Setup:**

````python
import pandas as pd

data = {
    "Artist":     ["Taylor Swift", "Drake", "BTS", "Ed Sheeran", "Billie Eilish", "The Weeknd"],
    "Genre":      ["Pop", "Hip-Hop", "K-Pop", "Pop", "Pop", "R&B"],
    "Streams_M":  [14200, 12800, 11500, 9700, 8900, 13100],
    "Albums":     [12, 9, 9, 6, 3, 7],
    "Year_Debut": [2006, 2009, 2013, 2004, 2019, 2010]
}

df = pd.DataFrame(data)
print(df)
````

### Exercise 1.1: Inspecting the DataFrame
Using `df`:
1. Print the first 3 rows
2. Print the shape — how many rows and columns?
3. Run `df.info()` — which columns are text and which are numbers?
4. Print just the `"Artist"` column
5. Print the `"Artist"` and `"Streams_M"` columns together

````python
# 1. First 3 rows
print(df.head(3))

# 2. Shape
print(df.shape)   # (6, 5)

# 3. Info
df.info()
# Artist       object  (text)
# Genre        object  (text)
# Streams_M    int64
# Albums       int64
# Year_Debut   int64

# 4. Single column
print(df["Artist"])

# 5. Two columns together
print(df[["Artist", "Streams_M"]])
````

### Exercise 1.2: Filtering Rows
Using `df`:
1. Filter to show only artists with `Streams_M` above 12000
2. Filter to show only **Pop** artists
3. Filter to show artists who debuted **before 2010** (`Year_Debut < 2010`)
4. **BONUS:** Filter to show **Pop** artists with more than 10000 streams

Print each result.

````python
# 1. Streams_M > 12000
print(df[ df["Streams_M"] > 12000 ])
# Taylor Swift (14200), Drake (12800) ... wait, also The Weeknd (13100)

# 2. Pop artists only
print(df[ df["Genre"] == "Pop" ])
# Taylor Swift, Ed Sheeran, Billie Eilish

# 3. Debuted before 2010
print(df[ df["Year_Debut"] < 2010 ])
# Taylor Swift (2006), Drake (2009), Ed Sheeran (2004)

# 4. BONUS: Pop AND Streams_M > 10000
print(df[ (df["Genre"] == "Pop") & (df["Streams_M"] > 10000) ])
# Taylor Swift (14200)
````

### Exercise 1.3: Quick Math
Using `df`:
1. What is the **average** number of streams (`Streams_M`)?
2. What is the **maximum** number of streams?
3. What is the **minimum** number of albums?
4. What is the **total** streams across all artists?
5. **BONUS:** What is the average streams for Pop artists only?

````python
# 1. Average streams
print("Average:", df["Streams_M"].mean())    # 11700.0

# 2. Maximum streams
print("Max:", df["Streams_M"].max())          # 14200

# 3. Minimum albums
print("Min albums:", df["Albums"].min())      # 3

# 4. Total streams
print("Total:", df["Streams_M"].sum())        # 70200

# 5. BONUS: Average streams for Pop only
pop = df[ df["Genre"] == "Pop" ]
print("Pop avg:", pop["Streams_M"].mean())    # 10933.33...
````

### Exercise 1.4: Sorting and Adding Columns
Using `df`:
1. Sort by `Streams_M` from **highest to lowest** and print
2. Sort by `Year_Debut` from **oldest to newest** and print
3. Add a column `"Streams_B"` = `Streams_M` divided by 1000 *(converts millions to billions)*
4. Add a column `"Is_Popular"` = `True` if `Streams_M >= 12000`, else `False`
5. Print `df[["Artist", "Streams_M", "Streams_B", "Is_Popular"]]`

````python
# 1. Sort by Streams_M descending
print(df.sort_values("Streams_M", ascending=False))

# 2. Sort by Year_Debut ascending (oldest first)
print(df.sort_values("Year_Debut"))

# 3. Add Streams_B column
df["Streams_B"] = df["Streams_M"] / 1000

# 4. Add Is_Popular column
df["Is_Popular"] = df["Streams_M"] >= 12000

# 5. Print selected columns
print(df[["Artist", "Streams_M", "Streams_B", "Is_Popular"]])
#           Artist  Streams_M  Streams_B  Is_Popular
# 0   Taylor Swift      14200       14.2        True
# 1          Drake      12800       12.8        True
# 2            BTS      11500       11.5       False
# 3     Ed Sheeran       9700        9.7       False
# 4  Billie Eilish       8900        8.9       False
# 5     The Weeknd      13100       13.1        True
````

## Part 2: Visualizations 📊

DataFrames have a built-in `.plot()` method. The key arguments are:
- `kind=` — type of chart: `"bar"`, `"barh"`, `"line"`, `"scatter"`
- `x=` — which column goes on the x-axis
- `y=` — which column goes on the y-axis

```python
df.plot(kind="bar", x="Artist", y="Streams_M")
```

**Setup:**

````python
import pandas as pd

data = {
    "Artist":     ["Taylor Swift", "Drake", "BTS", "Ed Sheeran", "Billie Eilish", "The Weeknd"],
    "Genre":      ["Pop", "Hip-Hop", "K-Pop", "Pop", "Pop", "R&B"],
    "Streams_M":  [14200, 12800, 11500, 9700, 8900, 13100],
    "Albums":     [12, 9, 9, 6, 3, 7],
    "Year_Debut": [2006, 2009, 2013, 2004, 2019, 2010]
}

df = pd.DataFrame(data)
````

### Challenge 1: Full Pandas Workflow
```python
data = {
    "Country":    ["USA", "China", "Japan", "Germany", "UK", "India"],
    "Region":     ["Americas", "Asia", "Asia", "Europe", "Europe", "Asia"],
    "GDP_T":      [25.5, 17.7, 4.2, 4.1, 3.1, 3.4],
    "Pop_M":      [335, 1412, 125, 84, 67, 1428],
    "Growth_pct": [2.1, 5.2, 1.9, 1.8, 4.1, 7.2]
}
```
1. Create `countries_df` and print its shape
2. Print the first 3 rows
3. Filter to show only **Asian** countries
4. What is the **average** `GDP_T`?
5. Sort by `Growth_pct` from **highest to lowest** and print
6. Add a column `"GDP_per_capita"` = `(GDP_T * 1_000_000) / Pop_M` *(GDP in millions per person)*
7. Print `countries_df[["Country", "GDP_T", "Pop_M", "GDP_per_capita"]]`

````python
import pandas as pd

data = {
    "Country":    ["USA", "China", "Japan", "Germany", "UK", "India"],
    "Region":     ["Americas", "Asia", "Asia", "Europe", "Europe", "Asia"],
    "GDP_T":      [25.5, 17.7, 4.2, 4.1, 3.1, 3.4],
    "Pop_M":      [335, 1412, 125, 84, 67, 1428],
    "Growth_pct": [2.1, 5.2, 1.9, 1.8, 4.1, 7.2]
}

# 1. Create and check shape
countries_df = pd.DataFrame(data)
print(countries_df.shape)   # (6, 5)

# 2. First 3 rows
print(countries_df.head(3))

# 3. Asian countries only
print(countries_df[ countries_df["Region"] == "Asia" ])
# China, Japan, India

# 4. Average GDP_T
print("Average GDP_T:", countries_df["GDP_T"].mean())   # 9.67

# 5. Sort by Growth_pct descending
print(countries_df.sort_values("Growth_pct", ascending=False))
# India (7.2), China (5.2), UK (4.1), ...

# 6. Add GDP_per_capita
countries_df["GDP_per_capita"] = (countries_df["GDP_T"] * 1_000_000) / countries_df["Pop_M"]

# 7. Print selected columns
print(countries_df[["Country", "GDP_T", "Pop_M", "GDP_per_capita"]])
````

### Challenge 2: Visualization Workflow
Using `countries_df` from Challenge 1:
1. Create a **horizontal bar chart** of `GDP_T` by `Country`
2. Create a **scatter plot** of `Pop_M` (x) vs `GDP_T` (y)
3. **BONUS:** Filter to only European countries, then create a bar chart of their `Growth_pct`

````python
# 1. Horizontal bar chart: GDP_T by Country
countries_df.plot(kind="barh", x="Country", y="GDP_T")

# 2. Scatter plot: Population vs GDP
countries_df.plot(kind="scatter", x="Pop_M", y="GDP_T")

# 3. BONUS: European countries only
europe = countries_df[ countries_df["Region"] == "Europe" ]
europe.plot(kind="bar", x="Country", y="Growth_pct")
````

### Challenge 3: Putting It All Together
```python
data = {
    "Movie":    ["Inception", "The Dark Knight", "Interstellar", "Parasite", "Dune", "Avatar"],
    "Genre":    ["Sci-Fi", "Action", "Sci-Fi", "Drama", "Sci-Fi", "Sci-Fi"],
    "Year":     [2010, 2008, 2014, 2019, 2021, 2009],
    "Rating":   [8.8, 9.0, 8.6, 8.5, 8.0, 7.9],
    "Revenue_M":[836, 1005, 773, 258, 401, 2923]
}
```
1. Create `movies_df` and print its shape
2. Filter to show only **Sci-Fi** movies
3. What is the **average** rating across all movies?
4. Sort by `Revenue_M` from **highest to lowest** — which movie earned the most?
5. Add a column `"Is_Blockbuster"` = `True` if `Revenue_M >= 800`
6. Create a **scatter plot** of `Rating` (x) vs `Revenue_M` (y)
7. Create a **bar chart** of `Rating` by `Movie`

````python
import pandas as pd

data = {
    "Movie":    ["Inception", "The Dark Knight", "Interstellar", "Parasite", "Dune", "Avatar"],
    "Genre":    ["Sci-Fi", "Action", "Sci-Fi", "Drama", "Sci-Fi", "Sci-Fi"],
    "Year":     [2010, 2008, 2014, 2019, 2021, 2009],
    "Rating":   [8.8, 9.0, 8.6, 8.5, 8.0, 7.9],
    "Revenue_M":[836, 1005, 773, 258, 401, 2923]
}

# 1. Create and check shape
movies_df = pd.DataFrame(data)
print(movies_df.shape)   # (6, 5)

# 2. Sci-Fi movies only
print(movies_df[ movies_df["Genre"] == "Sci-Fi" ])
# Inception, Interstellar, Dune, Avatar

# 3. Average rating
print("Average rating:", movies_df["Rating"].mean())   # 8.47

# 4. Sort by Revenue_M descending
print(movies_df.sort_values("Revenue_M", ascending=False))
# Avatar earned the most (2923M)

# 5. Add Is_Blockbuster
movies_df["Is_Blockbuster"] = movies_df["Revenue_M"] >= 800
print(movies_df[["Movie", "Revenue_M", "Is_Blockbuster"]])
# Inception, The Dark Knight, Avatar → True

# 6. Scatter plot: Rating vs Revenue
movies_df.plot(kind="scatter", x="Rating", y="Revenue_M")

# 7. Bar chart: Rating by Movie
movies_df.plot(kind="bar", x="Movie", y="Rating")
````
