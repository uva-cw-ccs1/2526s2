```python
# Task 3d: Write a function called classify_rating that takes a rating value and returns "low" if rating < 7.5, "medium" if 7.5 <= rating < 8.5, or "high" if rating >= 8.5. 

def classify_rating(rating_value):
    # rating_value should be a float or integer
    if rating_value < 7.5:
        result = "low"
    elif rating_value < 8.5:   # I don't have to check for rating_value >= 7.5 because the code will not see this line if rating_value is smaller than 7.5 anyway. 
        result = "medium"
    else:
        result = "high"
    return result

# Task 3d: Then apply it to the "rating" column to create a new column called "rating_label". Print films_df[["title", "rating", "acclaimed", "rating_label"]] to check your work. 

# Now the problem is the function we've written above expects a single rating_value but we are dealing with a dataframe. We need to feed the rating values one by one.  HERE COMES THE FOR LOOP!

label_list = [] 
for value in films_df['rating']:
    label = classify_rating(value)
    label_list.append(label)
print(label_list)

print ("--------------------")    # This makes it easier to see different data frames in the output

films_df["rating_label"] = label_list
print(films_df)

print ("--------------------")    # This makes it easier to see different data frames in the output

print(films_df[["title", "rating", "acclaimed", "rating_label"]])
```
