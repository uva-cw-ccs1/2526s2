### Challenge 1: Social Media Analytics
You're analyzing a social media post. Create variables:
- `post_likes = 340`
- `post_shares = 85`
- `post_comments = 52`

Calculate and print:
1. Total engagement (sum of all three)

### Challenge 2: Survey Responses
You collected survey responses: `[7, 8, 6, 9, 7, 8, 10, 6, 9, 8]`

Store this list in a variable called `responses`. Then:
1. Calculate the average response (sum divided by length)
2. Print: "Average rating: [average]"
3. Print the highest response
4. Print how many responses you collected

```python
responses = [7, 8, 6, 9, 7, 8, 10, 6, 9, 8]

average = sum(responses) / len(responses)
print('Average rating:', average)

print('Highest response:', max(responses))
print('Number of responses:', len(responses))
```

### Challenge 3: Student Data
Create these two lists:
- `student_names = ['Emma', 'Noah', 'Olivia', 'Liam']`
- `student_scores = [92, 87, 95, 89]`

Then:
1. Print the first student's name and score
2. Add a new student 'Ava' to the names list
3. Add her score of 91 to the scores list
4. Print both complete lists
5. Calculate and print the class average

```python
student_names = ['Emma', 'Noah', 'Olivia', 'Liam']
student_scores = [92, 87, 95, 89]

# 1. Print first student's name and score
print('First student:', student_names[0], 'Score:', student_scores[0])

# 2 & 3. Add new student
student_names.append('Ava')
student_scores.append(91)

# 4. Print both lists
print('Names:', student_names)
print('Scores:', student_scores)

# 5. Calculate class average
class_average = sum(student_scores) / len(student_scores)
print('Class average:', class_average)
```
