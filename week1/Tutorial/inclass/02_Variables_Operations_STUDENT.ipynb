## What are Variables?

Variables **store values** with a name.

**Python:** `age = 22` (store one value)

**SPSS:** Variables are columns in a dataset

Variables in Python are more flexible!

## Creating Variables

Format: `name = value`

```python
# Create a variable
age = 22
print(age)
```

```python
# Variables can hold any type
name = 'Emma'
rating = 4.5
print(name, rating)
```

## Using Variables in Operations

```python
followers = 1500
followers + 100  # Doesn't change followers!
```

```python
print(followers)  # Still 1500
```

```python
# To change it, reassign:
followers = followers + 100
print(followers)  # Now 1600
```

```python
# Shorthand: += means "add to itself"
followers += 50
print(followers)  # Now 1650
```

## Combining Variables

```python
# With numbers
likes = 500
comments = 50
total = likes + comments
print(total)
```

```python
# With strings
first = 'Emma'
last = 'Johnson'
full_name = first + ' ' + last
print(full_name)
```

```python
# Mix in print (use commas)
username = 'emma_j'
posts = 127
print(username, 'has', posts, 'posts')
```

## Variable Naming Rules

âœ… **Must:**
- Start with letter or underscore
- Use letters, numbers, underscores only
- No spaces

âœ… **Best practice:**
- Use descriptive names: `user_age` not `x`
- Use lowercase with underscores: `total_likes`

```python
# Good names
user_age = 25
total_followers = 1500
```

```python
# Bad names (will error!):
# 2cool = 20        # Can't start with number
# user age = 25     # No spaces
```

* Variables store values: `age = 22`

* Operations don't change variables unless you reassign
