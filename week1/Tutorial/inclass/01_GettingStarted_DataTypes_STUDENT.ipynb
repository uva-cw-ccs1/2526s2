## Data Types

Python has different types of data:

1. **Integers** - whole numbers: `42`
2. **Floats** - decimals: `3.14`
3. **Strings** - text (needs quotes): `'Hello'`
4. **Booleans** - True or False: `True`, `False`

```python
# Integer
22
```

```python
# Float
4.5
```

```python
# String (text needs quotes!)
'Hello World'
```

```python
# Boolean (True or False)
True
```

```python
# Check type with type()
type(3.14)
```

```python
# Booleans are useful for yes/no questions
is_verified = True
has_posts = False
print(is_verified, has_posts)
```
