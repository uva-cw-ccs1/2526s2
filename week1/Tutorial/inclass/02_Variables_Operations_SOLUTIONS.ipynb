````python
```xml
<VSCode.Cell language="markdown">
# Session 2: Variables & Operations

**Today:** Storing data in variables and using them in calculations
</VSCode.Cell>
<VSCode.Cell language="markdown">
## What are Variables?

Variables **store values** with a name.

**Python:** `age = 22` (store one value)

**SPSS:** Variables are columns in a dataset

Variables in Python are more flexible!
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Creating Variables

Format: `name = value`
</VSCode.Cell>
<VSCode.Cell language="python">
# Create a variable
age = 22
print(age)
</VSCode.Cell>
<VSCode.Cell language="python">
# Variables can hold any type
name = 'Emma'
rating = 4.5
print(name, rating)
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Using Variables in Operations
</VSCode.Cell>
<VSCode.Cell language="python">
followers = 1500
followers + 100  # Doesn't change followers!
</VSCode.Cell>
<VSCode.Cell language="python">
print(followers)  # Still 1500
</VSCode.Cell>
<VSCode.Cell language="python">
# To change it, reassign:
followers = followers + 100
print(followers)  # Now 1600
</VSCode.Cell>
<VSCode.Cell language="python">
# Shorthand: += means "add to itself"
followers += 50
print(followers)  # Now 1650
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Combining Variables
</VSCode.Cell>
<VSCode.Cell language="python">
# With numbers
likes = 500
comments = 50
total = likes + comments
print(total)
</VSCode.Cell>
<VSCode.Cell language="python">
# With strings
first = 'Emma'
last = 'Johnson'
full_name = first + ' ' + last
print(full_name)
</VSCode.Cell>
<VSCode.Cell language="python">
# Mix in print (use commas)
username = 'emma_j'
posts = 127
print(username, 'has', posts, 'posts')
</VSCode.Cell>
<VSCode.Cell language="markdown">
## Variable Naming Rules

âœ… **Must:**
- Start with letter or underscore
- Use letters, numbers, underscores only
- No spaces

âœ… **Best practice:**
- Use descriptive names: `user_age` not `x`
- Use lowercase with underscores: `total_likes`
</VSCode.Cell>
<VSCode.Cell language="python">
# Good names
user_age = 25
total_followers = 1500
</VSCode.Cell>
<VSCode.Cell language="python">
# Bad names (will error!):
# 2cool = 20        # Can't start with number
# user age = 25     # No spaces
</VSCode.Cell>
<VSCode.Cell language="markdown">
## ðŸŽ¯ Practice Exercise

Social media post analysis:
</VSCode.Cell>
<VSCode.Cell language="python">
# 1. Create: post_likes = 340, post_shares = 45, post_comments = 28
post_likes = 340
post_shares = 45
post_comments = 28
</VSCode.Cell>
<VSCode.Cell language="python">
# 2. Calculate total engagement (sum of all three)
total_engagement = post_likes + post_shares + post_comments
print(total_engagement)
</VSCode.Cell>
<VSCode.Cell language="python">
# 3. Create platform = 'Instagram'
platform = 'Instagram'
</VSCode.Cell>
<VSCode.Cell language="python">
# 4. Print: "Instagram post has 413 total engagements"
print(platform, 'post has', total_engagement, 'total engagements')
</VSCode.Cell>
<VSCode.Cell language="markdown">
## âœ… Key Takeaways

* Variables store values: `age = 22`
* Use `=` to assign, `==` to compare (later!)
* Operations don't change variables unless you reassign
* Shorthand: `x += 5` means `x = x + 5`
* Good names are descriptive and use underscores
* Print multiple values with commas: `print(name, age)`
</VSCode.Cell>
<VSCode.Cell language="markdown">

</VSCode.Cell>
```
````
