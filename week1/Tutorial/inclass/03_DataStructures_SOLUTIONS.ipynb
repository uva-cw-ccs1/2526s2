**Today:** Storing multiple values in ordered collections

## Lists: Ordered Collections

**List** = multiple items in order

Use square brackets `[ ]`

`````python
# Create lists
ages = [22, 25, 19, 31, 28]
platforms = ['Instagram', 'TikTok', 'Twitter']
print(ages)
print(platforms)
`````

## Accessing List Items

‚ö†Ô∏è **Python starts counting at 0!**

```
['Instagram', 'TikTok', 'Twitter']
      0          1         2
```

`````python
platforms = ['Instagram', 'TikTok', 'Twitter']

# First item (index 0)
print(platforms[0])
`````

`````python
# Last item (use -1)
print(platforms[-1])
`````

`````python
# Slice: get multiple items [start:end]
print(platforms[0:2])  # Get items 0 and 1 (not 2!)
`````

## Modifying Lists

`````python
likes = [100, 150, 200]

# Add to end
likes.append(250)
print(likes)
`````

`````python
# Change item
likes[0] = 120
print(likes)
`````

`````python
# Useful functions
numbers = [10, 50, 30, 70, 20]
print('Length:', len(numbers))
print('Sum:', sum(numbers))
print('Max:', max(numbers))
`````

## ‚úÖ Key Takeaways

* **Lists** `[item1, item2]` - ordered collections
* Indexing starts at 0: `my_list[0]`
* Negative indexing: `my_list[-1]` is last item
* Slicing: `my_list[start:end]` gets multiple items
* Add items: `my_list.append(item)`
* Change items: `my_list[0] = new_value`
* Useful functions: `len()`, `sum()`, `max()`
