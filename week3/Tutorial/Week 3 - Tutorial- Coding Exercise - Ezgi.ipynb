2- We use for loops to loop over a collection, one item at a time. 


```python
ccs1_class = ["Alen", "Maria", "Zoe", "Daniela D.", "Dorota", "Fenella", "Amelia", "Haoran", "Jeff", "Philip", "Uyen", "Jaden", "Alex", "Nathalie", "Mai Anh", "Ava", "Daniela Y.", "Mingxuan"]
# ccs1_class is a list, which holds the names of all students in CCS1 class this year. We need a collection to run a for loop.
print(type(ccs1_class))
```

```python
# I wonder, how many students are enrolled in the class... 
print(len(ccs1_class))
```

```python
# for loops always use the same structure "for variable_name in collection_name:"
for person in ccs1_class:
    print(person)  # this code is indented. That means, it will run when the last unindented code is activated. 
```

```python
check = "Ezgi" in ccs1_class # without the "for" keyword, "in" is just a keyword to check membership. Here, variable "check" holds the information whether "Ezgi" is in, therefore a member of ccs1_class. 
print(check)
```

```python
check = "Nathalie" in ccs1_class  #this line, by using the same variable name, overwrites the previous value of check.
print(check)
```

```python
check2 = (5==5)
print(check2)

check3 = (7==5)
print(check3)

check4 = (7!=5)   #You can use hashtag in a code cell to write a note inside your code. It's very useful to put reminders for yourself.
print(check4)
```

```python
#Going back to the for loop, let's print the names of students who have the letter "d" in their names. 
for student_name in ccs1_class:
    if "d" in student_name:  #this line is indented because we want to check whether the letter d is in the name for every student one by one. 
        print(student_name)  #this line is indented again because we want to run this code only if "d" is in the student_name.

#this code ignored Dorota and Danielas... Why? Please try to fix it. There are many solutions for this problem. Tip: .lower(), .upper()
```

```python
#let's have a list of students who don't have the letter "d" in their names. 
without_d = []

for student_name in ccs1_class:
    if ("d" in student_name) | ("D" in student_name): 
        continue  # you can use the keyword "continue" to tell your code to continue running without doing anything. If we did not write this here, if condition would give an error because it expects the next line to be indented.
    else:  #elif and else statements, following the if statement should be on the same vertical position with the if keyword. In other words, you don't indent elif and else statements further. 
        without_d.append(student_name)
print(without_d)

#You can fill without_d list another way. You don't need "continue" keyword or an else. Please try to do it in another way. 
```

3- range(x) is a function that we use with for loops to activate the for loop x times.


```python
x=5

for number in range(x):
    print(number, "hello")
```

```python
#We use range() because otherwise, if you put an integer as the last item of the for loop line, it will give an error. For loop needs a collection or a range of items to run. See below.
for number in 5:
    print(number)
```

```python
# You might also tell range where should it start counting and until which number. For instance, the code below prints all the years that Taylor Swift was alive.
for year in range(1989, 2027): #the last number will not be printed. It is the number range should STOP.
    print(year)
```

```python
# Did you have fun? If yes, let's make the python count these years two times! 
for yeeeaaaayy in range(2):
    for year in range(1989, 2027):
        print(year)
```

```python
# It was hard to see where the first loop ends.
for yeeeaaaayy in range(2):
    for year in range(1989, 2027):
        print(year)
    print("----------------------------------") #this will help. This code is indented once. That means it won't be affected by the second loop and will be activated after that. 
```

```python
#let's put these years in a list. 
taylors_memories = []

for year in range(1989, 2027): 
    taylors_memories.append(year)
print(taylors_memories)
```

```python
taylors_memories = []

for year in range(1989, 2027): 
    taylors_memories.append(year)
    print(taylors_memories)  #See how this line is indented? It means, this print function will be activated every time for loop iterates. 
print(taylors_memories) #See how this line is not indented? It means this print function will be activated only once after we are done with the for loop.
```

4- We use range function with index to navigate in lists. Because both range and lists start counting from zero.

```python
#Let's assume that you are curious about the tenth year Taylor Swift was alive in - for some reason... You can simply ask the list when it was. 
curious_number = 10
which_year = taylors_memories[curious_number]
print(which_year)  
```

```python
#If you count the years in the list above, you will see that 1999 is actually the 11th item in the list. That's because python starts counting from zero. So, for python, 1989 is the zeroth year Taylor Swift was alive, not the first. 
curious_number = 10
which_year = taylors_memories[curious_number -1]  #now, this is correct.
print(which_year)
```

```python
#Let's assume that you want to change the taylors_memories by changing the tenth year into a string memory.
print(taylors_memories[curious_number -1])   #this is 1998, we want to change it.

taylors_memories[curious_number-1] = "The year Titanic movie won many Oscars"
print(taylors_memories)
```

Now, here is a better example. You are working at a supermarket. We have a list for items, and another list for their price. A customer approaches with a shopping cart... 


```python
items = ["apple", "banana", "milk", "onion", "bread", "chocolate"]
price_list = [0.48, 0.76, 2.03, 0.12, 0.50, 3.25]

shopping_cart = ["apple", "apple", "banana", "chocolate", "chocolate", "apple", "bread"]  

# Let's calculate how much the customer should pay... 
```

```python
#The code below works, but it is a very badly written code. Because you have to do everything manually. 

cost = 0

for item in shopping_cart:
    if item == "apple":
        # add 0.48 to the cost. 
        cost = cost + 0.48
    elif item == "banana":   
        cost = cost + 0.76
    elif item == "milk":
        cost = cost + 2.03
    elif item == "onion":
        cost = cost + 0.12
    elif item == "bread": 
        cost = cost + 0.50
    elif item == "chocolate":
        cost = cost + 3.25
    else:
        print("Item is not on the list: " + item)

print(cost)
```

```python
# Now, let's assume that you cannot see the price_list or the item names. How would you write the code to calculate the cost? (Tip: use index)

```

```python
#Solution below

cost = 0

for item in shopping_cart:
    position = items.index(item)  #regarding this item from the shopping cart, what is its position in the "items" list? (get the index number and save that number in the variable "position")
    price = price_list[position]  #look at what is written at the same position in the "price_list" list (and save that in the variable "price")
    cost = cost + price   # cost += price  

#This is how you respect the privacy of your customers.
```
