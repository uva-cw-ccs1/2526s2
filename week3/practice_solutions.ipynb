### Exercise 1.9: Loops + Conditionals
Given: `prices = [45, 120, 30, 85, 150, 20, 95]`

Use a loop with conditionals to:
1. Count how many prices are "expensive" (>= 100)
2. Count how many prices are "moderate" (50-99)
3. Count how many prices are "cheap" (< 50)
4. Print all three counts

````python
prices = [45, 120, 30, 85, 150, 20, 95]

expensive = 0
moderate = 0
cheap = 0

for price in prices:
    if price >= 100:
        expensive += 1
    elif price >= 50:
        moderate += 1
    else:
        cheap += 1

print("Expensive:", expensive)  # 2
print("Moderate:", moderate)    # 2
print("Cheap:", cheap)          # 3
````

### Exercise 1.10: String Operations in Loops
Given: `tags = ['#python', '#DataScience', '#AI', '#MachineLearning', '#coding']`

Use a loop to:
1. Print only tags that contain "data" or "Data" (case-insensitive)
2. Count how many tags are longer than 8 characters

Hint: Use `.lower()` to make comparisons case-insensitive.

````python
tags = ['#python', '#DataScience', '#AI', '#MachineLearning', '#coding']

# Print tags containing "data"
for tag in tags:
    if "data" in tag.lower():
        print(tag)  # #DataScience

# Count tags longer than 8 characters
long_tags = 0
for tag in tags:
    if len(tag) > 8:
        long_tags += 1

print("Tags longer than 8 characters:", long_tags)  # 3
````

### Exercise 2.5: Looping Through Dictionaries
Create this dictionary:
```python
inventory = {
    "apples": 50,
    "bananas": 75,
    "oranges": 30,
    "grapes": 60
}
```

Then:
1. Use a loop to print each fruit name
2. Use a loop to calculate the total quantity of all fruits
3. Use a loop to print items with quantity >= 50

````python
inventory = {
    "apples": 50,
    "bananas": 75,
    "oranges": 30,
    "grapes": 60
}

# Print each fruit name
for fruit in inventory:
    print(fruit)

# Calculate total quantity
total = 0
for quantity in inventory.values():
    total += quantity
print("Total quantity:", total)  # 215

# Print items with quantity >= 50
for fruit, quantity in inventory.items():
    if quantity >= 50:
        print(f"{fruit}: {quantity}")
````

### Exercise 2.6: Dictionary with Loops and Conditionals
Create this dictionary:
```python
exam_scores = {
    "Emma": 92,
    "Noah": 87,
    "Olivia": 95,
    "Liam": 78,
    "Ava": 88
}
```

Use a loop to:
1. Print students who scored >= 90
2. Calculate the class average
3. Count how many students passed (>= 80)

````python
exam_scores = {
    "Emma": 92,
    "Noah": 87,
    "Olivia": 95,
    "Liam": 78,
    "Ava": 88
}

# Print students with >= 90
print("High achievers:")
for student, score in exam_scores.items():
    if score >= 90:
        print(f"{student}: {score}")

# Calculate class average
average = sum(exam_scores.values()) / len(exam_scores)
print(f"\nClass average: {average}")  # 88.0

# Count students who passed
passed = 0
for score in exam_scores.values():
    if score >= 80:
        passed += 1
print(f"Students passed: {passed}")  # 4
````

### Exercise 2.7: Creating Dictionaries from Lists
You have two parallel lists:
```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]
prices = [999, 25, 75, 350]
```

Create a dictionary called `product_catalog` where:
- Keys are product names
- Values are prices

Print the resulting dictionary.

````python
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]
prices = [999, 25, 75, 350]

# Solution 1: Using a loop
product_catalog = {}
for i in range(len(products)):
    product_catalog[products[i]] = prices[i]

print(product_catalog)
# {'Laptop': 999, 'Mouse': 25, 'Keyboard': 75, 'Monitor': 350}

# Solution 2: Using zip (more Pythonic)
# product_catalog = dict(zip(products, prices))
# print(product_catalog)
````

### Challenge 1: Movie Database
Create a list of 4 movie dictionaries. Each should have:
- "title": movie name
- "year": release year
- "rating": score out of 10
- "watched": boolean

Then:
1. Print titles of movies you've watched
2. Calculate the average rating of all movies
3. Find and print movies from 2020 or later
4. Update the "watched" status of one movie to True

````python
movies = [
    {"title": "Inception", "year": 2010, "rating": 8.8, "watched": True},
    {"title": "Dune", "year": 2021, "rating": 8.0, "watched": False},
    {"title": "The Matrix", "year": 1999, "rating": 8.7, "watched": True},
    {"title": "Oppenheimer", "year": 2023, "rating": 8.5, "watched": False}
]

# 1. Print watched movies
print("Watched movies:")
for movie in movies:
    if movie["watched"]:
        print(movie["title"])

# 2. Calculate average rating
total_rating = sum(movie["rating"] for movie in movies)
avg_rating = total_rating / len(movies)
print(f"\nAverage rating: {avg_rating}")  # 8.5

# 3. Print movies from 2020 or later
print("\nRecent movies (2020+):")
for movie in movies:
    if movie["year"] >= 2020:
        print(f"{movie['title']} ({movie['year']})")

# 4. Update watched status
movies[1]["watched"] = True
print(f"\nUpdated Dune status: {movies[1]['watched']}")
````

### Challenge 2: E-commerce Analysis
You have this shopping cart data:
```python
cart = [
    {"item": "Laptop", "price": 999, "quantity": 1, "category": "Electronics"},
    {"item": "Mouse", "price": 25, "quantity": 2, "category": "Electronics"},
    {"item": "Notebook", "price": 5, "quantity": 5, "category": "Stationery"},
    {"item": "USB Cable", "price": 10, "quantity": 3, "category": "Electronics"}
]
```

Calculate:
1. Total cost (price √ó quantity for each item, then sum)
2. Total number of items in cart
3. Most expensive single item
4. Total spent on "Electronics" category

````python
cart = [
    {"item": "Laptop", "price": 999, "quantity": 1, "category": "Electronics"},
    {"item": "Mouse", "price": 25, "quantity": 2, "category": "Electronics"},
    {"item": "Notebook", "price": 5, "quantity": 5, "category": "Stationery"},
    {"item": "USB Cable", "price": 10, "quantity": 3, "category": "Electronics"}
]

# 1. Total cost
total_cost = 0
for item in cart:
    total_cost += item["price"] * item["quantity"]
print(f"Total cost: ${total_cost}")  # $1104

# 2. Total number of items
total_items = 0
for item in cart:
    total_items += item["quantity"]
print(f"Total items: {total_items}")  # 11

# 3. Most expensive single item
max_price = 0
most_expensive = ""
for item in cart:
    if item["price"] > max_price:
        max_price = item["price"]
        most_expensive = item["item"]
print(f"Most expensive item: {most_expensive} (${max_price})")  # Laptop ($999)

# 4. Total spent on Electronics
electronics_total = 0
for item in cart:
    if item["category"] == "Electronics":
        electronics_total += item["price"] * item["quantity"]
print(f"Electronics total: ${electronics_total}")  # $1079
````

### Challenge 3: Social Media Analytics
Create a dictionary to track engagement data:
```python
posts = {
    "post1": {"likes": 150, "comments": 23, "shares": 12},
    "post2": {"likes": 89, "comments": 15, "shares": 5},
    "post3": {"likes": 203, "comments": 45, "shares": 28},
    "post4": {"likes": 67, "comments": 8, "shares": 3}
}
```

Calculate:
1. Total engagement for each post (likes + comments + shares)
2. Which post has the highest engagement?
3. Average number of likes across all posts
4. How many posts have more than 100 likes?

````python
posts = {
    "post1": {"likes": 150, "comments": 23, "shares": 12},
    "post2": {"likes": 89, "comments": 15, "shares": 5},
    "post3": {"likes": 203, "comments": 45, "shares": 28},
    "post4": {"likes": 67, "comments": 8, "shares": 3}
}

# 1. Total engagement for each post
print("Engagement per post:")
for post_id, data in posts.items():
    total = data["likes"] + data["comments"] + data["shares"]
    print(f"{post_id}: {total}")

# 2. Highest engagement
max_engagement = 0
top_post = ""
for post_id, data in posts.items():
    total = data["likes"] + data["comments"] + data["shares"]
    if total > max_engagement:
        max_engagement = total
        top_post = post_id
print(f"\nTop post: {top_post} with {max_engagement} total engagement")

# 3. Average likes
total_likes = 0
for data in posts.values():
    total_likes += data["likes"]
avg_likes = total_likes / len(posts)
print(f"Average likes: {avg_likes}")  # 127.25

# 4. Posts with > 100 likes
high_like_count = 0
for data in posts.values():
    if data["likes"] > 100:
        high_like_count += 1
print(f"Posts with >100 likes: {high_like_count}")  # 2
````
