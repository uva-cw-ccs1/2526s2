### Exercise 1.2: Inspecting a DataFrame
Run the **setup cell** below first, then answer the questions.

## Part 2: Working with Data 🛠️

In real life you won't type data by hand — it lives in files. And once you have it, you'll filter, sort, and transform it.

**Run the setup cell below before starting exercises 2.2–2.6.**

**Setup:**

````python
import pandas as pd

data = {
    "Name":   ["Emma", "Noah", "Olivia", "Liam", "Sophia", "James"],
    "Age":    [20, 22, 19, 21, 20, 23],
    "Major":  ["Economics", "Psychology", "Economics", "History", "Psychology", "Economics"],
    "Score":  [88, 75, 92, 68, 85, 79],
    "Passed": [True, True, True, False, True, True]
}

df = pd.DataFrame(data)
print(df)
````

### Exercise 2.1: Reading and Writing CSV

1. Save `df` to a file called `"my_students.csv"` (no index)
2. Load it back into a new variable called `df_from_csv`
3. Print `df_from_csv`
4. Check its shape — it should be `(6, 5)`

````python
# 1. Save to CSV
df.to_csv("my_students.csv", index=False)

# 2. Load back
df_from_csv = pd.read_csv("my_students.csv")

# 3. Print
print(df_from_csv)

# 4. Check shape
print(df_from_csv.shape)   # (6, 5)
````

### Exercise 2.2: Filtering Rows
Using `df`:
1. Filter to show only students who **Passed** (`Passed == True`)
2. Filter to show only students aged **21 or older**
3. Filter to show only **Economics** students

Print each filtered result.

````python
# 1. Students who passed
print(df[ df["Passed"] == True ])

# 2. Students aged 21 or older
print(df[ df["Age"] >= 21 ])

# 3. Economics students only
print(df[ df["Major"] == "Economics" ])
#      Name  Age      Major  Score  Passed
# 0    Emma   20  Economics     88    True
# 2  Olivia   19  Economics     92    True
# 5   James   23  Economics     79    True
````

### Exercise 2.3: Filtering with Multiple Conditions
Using `df`:
1. Filter to show students with `Score >= 80` **AND** who Passed
2. Filter to show students who study **Economics OR Psychology**
   *(Hint: use `|` for OR, wrap each condition in parentheses)*
3. Filter to show **Economics** students with `Score > 80`

````python
# 1. Score >= 80 AND Passed
print(df[ (df["Score"] >= 80) & (df["Passed"] == True) ])

# 2. Economics OR Psychology
print(df[ (df["Major"] == "Economics") | (df["Major"] == "Psychology") ])

# 3. Economics students with Score > 80
print(df[ (df["Major"] == "Economics") & (df["Score"] > 80) ])
#     Name  Age      Major  Score  Passed
# 0   Emma   20  Economics     88    True
# 2 Olivia   19  Economics     92    True
````

### Exercise 2.4: Quick Math
Using `df`:
1. What is the **average** Score?
2. What is the **highest** Score?
3. What is the **lowest** Age?
4. What is the **total** of all Scores added together?
5. **BONUS:** What is the average Score of students who Passed?

````python
# 1. Average Score
print("Average:", df["Score"].mean())   # 81.17

# 2. Highest Score
print("Highest:", df["Score"].max())    # 92

# 3. Lowest Age
print("Lowest age:", df["Age"].min())   # 19

# 4. Sum of all Scores
print("Total:", df["Score"].sum())      # 487

# 5. BONUS: Average score of students who passed
passed = df[ df["Passed"] == True ]
print("Avg (passed):", passed["Score"].mean())   # 83.8
````

### Exercise 2.5: Sorting
Using `df`:
1. Sort students by Score from **highest to lowest** and print
2. Sort students by Age from **youngest to oldest** and print
3. **BONUS:** Sort by Major A to Z, save the result as `df_sorted`, then print it

````python
# 1. Sort by Score descending
print(df.sort_values("Score", ascending=False))

# 2. Sort by Age ascending (youngest first)
print(df.sort_values("Age"))

# 3. BONUS: Sort by Major A to Z
df_sorted = df.sort_values("Major")
print(df_sorted)
````

### Exercise 2.6: Adding New Columns
Using `df`:
1. Add a column `"Score_pct"` = Score divided by 100 *(e.g. 88 → 0.88)*
2. Add a column `"Is_Top"` that is `True` if Score >= 85, else `False`
3. Print `df[["Name", "Score", "Score_pct", "Is_Top"]]`

### Challenge 1: Explore a New Dataset
```python
data = {
    "Product":  ["Laptop", "Mouse", "Keyboard", "Headphones", "Webcam", "Monitor"],
    "Price":    [999, 25, 75, 120, 60, 350],
    "Stock":    [10, 200, 150, 80, 45, 20]
}
```
1. Create `products_df` from the data above
2. Print its shape
3. Filter to show only products priced **under 100**
4. What is the **average** price?
5. Sort by Price from **cheapest to most expensive** and print

````python
import pandas as pd

data = {
    "Product":  ["Laptop", "Mouse", "Keyboard", "Headphones", "Webcam", "Monitor"],
    "Price":    [999, 25, 75, 120, 60, 350],
    "Stock":    [10, 200, 150, 80, 45, 20]
}

# 1. Create DataFrame
products_df = pd.DataFrame(data)

# 2. Shape
print(products_df.shape)   # (6, 3)

# 3. Products under $100
print(products_df[ products_df["Price"] < 100 ])

# 4. Average price
print("Average price:", products_df["Price"].mean())   # 271.5

# 5. Sort cheapest first
print(products_df.sort_values("Price"))
````

### Challenge 2: Data Transformation
Using `products_df` from Challenge 1:
1. Add a column `"Total_Value"` = Price × Stock *(total inventory value in $)*
2. Add a column `"Is_Expensive"` = `True` if Price > 200, else `False`
3. Print `products_df[["Product", "Price", "Total_Value", "Is_Expensive"]]`
4. Filter to show only products where `Total_Value > 5000`

````python
# 1. Total inventory value
products_df["Total_Value"] = products_df["Price"] * products_df["Stock"]

# 2. Is_Expensive flag
products_df["Is_Expensive"] = products_df["Price"] > 200

# 3. Print selected columns
print(products_df[["Product", "Price", "Total_Value", "Is_Expensive"]])

# 4. High-value products (Total_Value > 5000)
print(products_df[ products_df["Total_Value"] > 5000 ])
#    Product  Price  Stock  Total_Value  Is_Expensive
# 2 Keyboard     75    150        11250         False
# 3 Headphones  120    80         9600         False
````

### Challenge 3: Putting It All Together
```python
data = {
    "City":       ["London", "Paris", "Tokyo", "Sydney", "New York", "Berlin"],
    "Country":    ["UK", "France", "Japan", "Australia", "USA", "Germany"],
    "Population": [9.0, 2.2, 14.0, 5.3, 8.3, 3.7],
    "Area_km2":   [1572, 105, 2194, 12368, 783, 892]
}
```
1. Create `cities_df` and print its shape
2. Print the first 3 rows
3. Filter to show cities with population **above 5 million**
4. What is the **highest** population value? *(use `.max()`)*
5. Add a column `"Density"` = `(Population × 1_000_000) / Area_km2` — people per km²
6. Sort by `Density` from highest to lowest and print `cities_df[["City", "Population", "Density"]]`

````python
import pandas as pd

data = {
    "City":       ["London", "Paris", "Tokyo", "Sydney", "New York", "Berlin"],
    "Country":    ["UK", "France", "Japan", "Australia", "USA", "Germany"],
    "Population": [9.0, 2.2, 14.0, 5.3, 8.3, 3.7],
    "Area_km2":   [1572, 105, 2194, 12368, 783, 892]
}

# 1. Create and check shape
cities_df = pd.DataFrame(data)
print(cities_df.shape)   # (6, 4)

# 2. First 3 rows
print(cities_df.head(3))

# 3. Cities with population > 5 million
print(cities_df[ cities_df["Population"] > 5 ])

# 4. Highest population
print("Max population:", cities_df["Population"].max())   # 14.0 (Tokyo)

# 5. Add Density column
cities_df["Density"] = (cities_df["Population"] * 1_000_000) / cities_df["Area_km2"]

# 6. Sort by Density descending
print(cities_df[["City", "Population", "Density"]].sort_values("Density", ascending=False))
# Paris is the densest!
````
