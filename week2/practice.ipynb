## Part 4: Counter Variables

A **counter** is a variable (usually starting at 0) that you increment inside a loop using `+= 1`.

```python
count = 0
for item in my_list:
    count += 1  # same as count = count + 1
print(count)
```

## Part 5: `while` Loops

A `while` loop keeps running **as long as a condition is `True`**.

```python
while condition:
    # this code repeats
```

**Be careful:** if the condition never becomes `False`, the loop runs forever! Always make sure something inside the loop changes the condition.
