## Part 2: Conditional Statements

**Task 5a:** Temperature check.

```python
temperature = 22

if temperature > 30:
    print("It's hot!")
elif temperature >= 15:
    print("It's nice outside.")
else:
    print("It's cold!")

# Output: It's nice outside.
```

**Task 5b:** Weather conditional.

```python
weather = "rainy"

if weather == "rainy":
    print("Take an umbrella!")
elif weather == "sunny":
    print("Wear sunglasses!")
elif weather == "snowy":
    print("Wear a warm coat!")
else:
    print("Just wear a hat.")

# Output: Take an umbrella!
```

**Task 5c:** Rainy AND sunny â€” what happens?

```python
rainy = True
sunny = True

if rainy:
    print("Take an umbrella!")
elif sunny:
    print("Wear sunglasses!")
else:
    print("Just wear a hat.")

# Output: Take an umbrella!
# Python is a lazy reader! It found that "rainy" is True, so it runs
# that block and skips everything else. It never even checks "sunny".
```

**Task 5d:** Fix it so both rainy AND sunny is handled.

```python
rainy = True
sunny = True

if rainy and sunny:
    print("Wear a raincoat!")
elif rainy:
    print("Take an umbrella!")
elif sunny:
    print("Wear sunglasses!")
else:
    print("Just wear a hat.")

# Output: Wear a raincoat!
# The combined condition is checked FIRST, before the individual ones.
```

### Part 3 (continued): Combining Loops and Conditionals

```python
movies = ["Gladiator", "The Matrix", "The Lord of the Rings", "Citizen Kane", "Avatar 2", "Inside Out 2", "Jaws"]
year = [2000, 1999, 2001, 1941, 2022, 2024, 1975]
score = [8.5, 8.7, 8.9, 8.2, 7.5, 7.5, 8.1]
is_watched = [True, True, True, False, True, False, False]
```

**Task 8a:** Print every movie with its year.

```python
for i in range(len(movies)):
    print(movies[i] + " (" + str(year[i]) + ")")
```

**Task 8b:** Print movies released before 2000.

```python
for i in range(len(movies)):
    if year[i] < 2000:
        print(movies[i])

# Output:
# The Matrix
# Citizen Kane
# Jaws
```

**Task 8c:** Find a movie containing `"Rings"`.

```python
for movie in movies:
    if "Rings" in movie:
        print(movie)

# Output: The Lord of the Rings
```

**Task 8d:** Sort movies into must_watch and optional lists.

```python
must_watch = []
optional = []

for i in range(len(movies)):
    if score[i] >= 8.3:
        must_watch.append(movies[i])
    else:
        optional.append(movies[i])

print("Must watch:", must_watch)
print("Optional:", optional)

# Must watch: ['Gladiator', 'The Matrix', 'The Lord of the Rings']
# Optional: ['Citizen Kane', 'Avatar 2', 'Inside Out 2', 'Jaws']
```

## Part 4: Counter Variables

**Task 9a:** Count how many movies have been watched.

```python
watched_count = 0

for i in range(len(movies)):
    if is_watched[i] == True:  # or simply: if is_watched[i]:
        watched_count += 1

print("Movies watched:", watched_count)  # Output: 4
```

**Task 9b:** Count movies with score above 8.0.

```python
count = 0

for s in score:
    if s > 8.0:
        count += 1

print("Movies with score above 8.0:", count)  # Output: 5
```

**Task 11:** Print every drink and snack combination.

```python
drinks = ["coffee", "tea", "juice"]
snacks = ["cookies", "cake", "chips"]

for drink in drinks:
    for snack in snacks:
        print(drink + " and " + snack)
```

## Part 7: Putting It All Together

**Task 12a:** Decrease all scores by 1.

```python
new_scores = []

for s in score:
    new_scores.append(s - 1)

print(new_scores)  # [7.5, 7.7, 7.9, 7.2, 6.5, 6.5, 7.1]
```

**Task 12b:** Add 2 new movies to all lists.

```python
movies.append("Inception")
year.append(2010)
score.append(8.8)
is_watched.append(True)

movies.append("Parasite")
year.append(2019)
score.append(8.5)
is_watched.append(True)

print(movies)
```

**Task 12c (Challenge):** Find the highest-rated movie.

```python
highest_score = score[0]
highest_index = 0

for i in range(len(score)):
    if score[i] > highest_score:
        highest_score = score[i]
        highest_index = i

print("Highest rated movie:", movies[highest_index], "with a score of", highest_score)
```
