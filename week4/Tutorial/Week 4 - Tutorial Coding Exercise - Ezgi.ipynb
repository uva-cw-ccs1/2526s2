## The question that I forgot to answer in the last course... Enumerate() function is a function that you DON'T NEED TO LEARN - because you can do the same thing by using index, counter variables and loops.

```python
my_flowers = ["tulip", "orchid", "poppy", "lavendar", "lotus", "daffodil", "jasmine", "dahlia", "bellflower"]
```

```python
#enumerate() function allows you to "call" and print both the index and the item of a list, together, when used wih the for loop. 
for i, flower in enumerate(my_flowers):
    print(i)
    print(flower)
    print(i, flower)
    print("------")
```

```python
# first alternative to enumerate:
for flower in my_flowers:
    print(my_flowers.index(flower))
    print(flower)
    print(my_flowers.index(flower), flower)    # print(str(my_flowers.index(flower)) + " " + flower)
    print("------")
```

```python
# second alternative to enumerate, if you cannot remember how to write list_name.index(item_name)
index = 0
for flower in my_flowers:
    print(index)
    print(flower)
    print(str(index) + " " + flower)       # print(index, flower)
    print("------")
    index = index + 1       # index += 1
    # Don't forget to add the counter variable! 
```

```python
# You can also initialize index from -1 and print exactly the same output with the codes above by just changing the position of the counter variable
index = -1
for flower in my_flowers:
    index = index + 1       # index += 1
    print(index)
    print(flower)
    print(str(index) + " " + flower)       # print(index, flower)
    print("------")
```

### As you can see, there are MANY ways to come up with exactly the same output. As long as you know the basics of Python, you can always figure it out yourself. It's like a puzzle!

# Dictionaries

Dictionaries are variables that can store multiple values, just like lists. However, they allow you to store your data with "keys" and make searching for a specific information much more easier!

# Functions

### Functions allow you to write a code once and use it multiple times without writing it over and over again. Basically, if you are copying and pasting a code that you write to change only a small part of it, you can probably write a function instead.

```python
# Remember this? 

super_heroes = ["Batman", "Superman", "Catwoman", "Spiderman", "Daredevil", "Invisible Woman", "Storm"]

real_identity = ["Bruce Wayne", "Clark Kent", "Selina Kyle", "Peter Parker", "Matt Murdock", "Sue Storm", "Ororo Munroe"]

comic_appearances = [24000, 20000, 5000, 15000, 500, 1500, 3000]

criminals_arrested = [12930, 12048, 94, 9824, 13012, 8721, 202] 

audience_ratings = [9.8, 8.5, 9.2, 9.7, 8.2, 8.4, 8.7]

is_there_upcoming_TVserie = [True, False, False, True, True, False, True]



# The code below is always the same. It's boring to write this over and over again.

real_identity_record = dict(zip(super_heroes, real_identity))
print(real_identity_record)

comic_record = dict(zip(super_heroes, comic_appearances))
print(comic_record)

criminal_record = dict(zip(super_heroes, criminals_arrested))
print(criminal_record)

audience_record = dict(zip(super_heroes, audience_ratings))
print(audience_record)

Tv_series_record = dict(zip(super_heroes, is_there_upcoming_TVserie))
print(Tv_series_record)
```
